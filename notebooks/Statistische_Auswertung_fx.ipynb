### K-Means Modellierung

Die vorangegangen Untersuchungen zeigen, dass zwar zwischen einigen Variablen hohe Korrelationen zu beoachten sind, zwischen den restlichen Variablen sind jedoch die korrelationen recht gering.  weitere Möglichkeit um Zusammenhänge zwischen den Daten aufzudecken besteht in der Bestimmung von clustern. Dabei werden cluster um K Mittelwerte bestimmt.
Im vorangegangen war eine mäßige negative Korrelation zwischen der Breite und der Länge der Fehlerstellen festgestellt worden. Um die genauere Abhängigkeit der Daten von den beiden Variablen zu utersuchen, wirde die k-means Modellierung verwendet, um Clsuter von Datenpunkte um Werte von Breite und Länge von den Felherstelle nzu finden. 
Dazu werden die beiden Variablen skalliert und transformiert, so dass sich ein Array von Tupeln (X,Y) ergeben.

Um die optimale Anzahl k von Clsuter Zentren zu bestimmen, wird die k-Means Cluster Analyse schrittweise von K=1 bis 11 druchgefuhrt und der Wert für SSE bestimmt. Dabei ist ein deutlicher Abfall von SSE bis k=11 zu erkennen. Dabei ist zu erkennen, dass die Verbesserung der Ergebnissen immer kleienr wird. Bei k=100 ist der Wert verhältnismäßoig gut. 

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['X', 'Y']])
scaled_features
```

```python
kmeans = KMeans(
     init="random",
     n_clusters=3,
     n_init=10,
     max_iter=300,
     random_state=42
 )
```

```python
kmeans.fit(scaled_features)
```

```python
# The lowest SSE value
kmeans.inertia_
```

```python
# Final locations of the centroid
kmeans.cluster_centers_
```

```python
# The number of iterations required to converge
kmeans.n_iter_
```

```python
kmeans.labels_
```

Um die optimale Anzahl k von Clsuter Zentren zu bestimmen, wird die k-Means Cluster Analyse schrittweise von K=1 bis 11 druchgefuhrt und der Wert für SSE bestimmt. Dabei ist ein deutlicher Abfall von SSE bis k=11 zu erkennen. Dabei ist zu erkennen, dass die Verbesserung der Ergebnissen immer kleienr wird. Bei k=10 ist der Wert verhältnismäßig gut. 
Mit Hilfe der Siloutéten Analse läßt sich weiterhin die Güte der clsuter bestimmen. Die Größe gbit an wie weit Pnkt in unterschiedlichen Clustern voneinander entfernt sind vergleichen mit dem Absand von Punkten innherhablb des clsuters. Desto größer der Wert der Sioutten analse also ist, desto besser sind die aufgeteilten cluster. Es sit erkennbar, dass der Wert mit steigender Clusteranzahl erst stark steigt und dann ab ca. 5 Clstern sich kaum noch ändert. Daher sind 4-5 Clsuter eine optimale Einteilung. 

```python
kmeans_kwargs = {
     "init": "random",
     "n_init": 10,
     "max_iter": 300,
     "random_state": 42,
 }

# A list holds the SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(
    range(1, 11), sse, curve="convex", direction="decreasing"
)
kl.elbow
```

```python
# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(scaled_features)
    score = silhouette_score(scaled_features, kmeans.labels_)
    silhouette_coefficients.append(score)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```
